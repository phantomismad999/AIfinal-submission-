**Missing Values: Drop or Impute**

The dropna () method was used to drop rows with missing data. This option was selected since incomplete records may create a mistake and result in incorrect findings when it comes to the analysis and visualization. Because the dataset is not very small, and there were enough data points after eliminating these rows, the process of dropping the missing values did not have a significant impact on the entire analysis.

The use of imputation was avoided since the assumption in filling the missing values would imply bias and would lower the credibility of the findings.

Composite Development Score:

```python
# 1. Create Composite Score
hdi_south_asia['Composite_Score'] = (
    0.30 * hdi_south_asia['life_expectancy'] +
    0.30 * hdi_south_asia['gross_inc_percap']
)
hdi_south_asia[['country', 'life_expectancy', 'gross_inc_percap', 'Composite_Score']]
```

```python
#Rank South Asian countries based on Composite Score.
hdi_south_asia = hdi_south_asia.sort_values(by='Composite_Score', ascending=False)
hdi_south_asia
```

```python
# Create ranks
hdi_south_asia['Composite_Rank'] = hdi_south_asia['Composite_Score'].rank(ascending=False, method='min')
hdi_south_asia['HDI_Rank'] = hdi_south_asia['hdi'].rank(ascending=False, method='min')
hdi_south_asia['Rank_Difference'] = hdi_south_asia['HDI_Rank'] - hdi_south_asia['Composite_Rank']

comparison = hdi_south_asia[['country', 'Composite_Score', 'Composite_Rank', 'hdi', 'HDI_Rank', 'Rank_Difference']]
comparison_sorted =comparison.sort_values(by='HDI_Rank')
print(comparison_sorted)
```

```python
# Function to detect outliers

def detect_outliers(column_data):
    Q1 = column_data.quantile(0.25)
    Q3 = column_data.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (column_data < lower) | (column_data > upper)

# Detect outliers in HDI and GNI per Capita
hdi_south_asia['HDI_Outlier'] = detect_outliers(hdi_south_asia['hdi'])
hdi_south_asia['GNI_Outlier'] = detect_outliers(hdi_south_asia['gross_inc_percap'])

# Flag any outlier
hdi_south_asia['Any_Outlier'] = hdi_south_asia['HDI_Outlier'] | hdi_south_asia['GNI_Outlier']

normal = hdi_south_asia[~hdi_south_asia['Any_Outlier']]
outliers = hdi_south_asia[hdi_south_asia['Any_Outlier']]

# Display outlier flags
hdi_south_asia[['country', 'hdi', 'gross_inc_percap', 'HDI_Outlier', 'GNI_Outlier', 'Any_Outlier']]
```

```python
plt.figure(figsize=(8,6))
plt.scatter(normal['gross_inc_percap'], normal['hdi'], color='blue', label='Normal', s=20)
plt.scatter(outliers['gross_inc_percap'], outliers['hdi'], color='red', label='Outlier')
plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.title('GNI per Capita vs HDI (Outliers Highlighted)')
plt.legend()
plt.show()
```

```python
#Compute Pearson correlation of each metric with HDI
corr_gender = hdi_south_asia["gender_development"].corr(hdi_south_asia["hdi"])
corr_life = hdi_south_asia["life_expectancy"].corr(hdi_south_asia["hdi"])

print("Correlation (Gender Development vs HDI):", corr_gender)
print("Correlation (Life Expectancy vs HDI):", corr_life)
```

Outlier Detection:


```python
#Define a function to detect outliers using 1.5 × IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    return outliers
```
